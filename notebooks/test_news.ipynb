```python
from dotenv import load_dotenv
import requests
import json
import os
```

```python
load_dotenv()
```

Checking this source of news: https://www.alphavantage.co/documentation/#intelligence

```python
api_key = os.environ["AV_API_KEY"]
url = f'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers=BTC&apikey={api_key}'
r = requests.get(url)
data = r.json()

print(json.dumps(data, indent=4))
```
